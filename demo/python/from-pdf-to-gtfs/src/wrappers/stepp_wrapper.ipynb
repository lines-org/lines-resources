```python
import zipfile
import geopandas as gpd
```

```python
zip_path = '../resources/aveiro-bus.stepp.zip'
ext_path = '../resources/aveiro-bus.stepp'
gdb_path = '../resources/aveiro-bus.stepp/dados/dados.gdb'
```

```python
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(ext_path)
```

```python
# Load the layers
circulacoes = gpd.read_file(gdb_path, layer='circulacoes')
paragens = gpd.read_file(gdb_path, layer='paragens')
segmentos = gpd.read_file(gdb_path, layer='segmentos')
servicos = gpd.read_file(gdb_path, layer='servicos')

# Convert coordinates to EPSG:4326
paragens.to_crs(epsg=4326, inplace=True)
segmentos.to_crs(epsg=4326, inplace=True)
```

```python
print(type(circulacoes))
print(type(paragens))
print(type(segmentos))
print(type(servicos))
```
